```python
import sys
print(sys.version)
#For this notebook to work, Python must be 3.6.4 or 3.6.5

import numpy as np
import pandas as pd
from IPython.display import display

from plotnine import *
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import Data Set
cancer = pd.read_csv('data/breast-cancer-wisconsin-data/data.csv', sep=',')
```

```python
#Inspect data set
display(cancer[:][:5])
print(cancer.shape)
```

```python
#Check data types 
cancer.dtypes
```

```python
# Divide the data into X and y (output, labels)
X = cancer.iloc[:, 2:].values
y_categorical = cancer.iloc[:, 1].values
```

```python
# Sanity check
display(X[:][:5])
print(X.shape)
```

```python
# We can see that # 19, 20 and 21 are Benign within the array that goes from 0:29
display(y_categorical[:30])
print(y_categorical.shape)
```

```python
# Encoding categorical data
from sklearn.preprocessing import LabelEncoder
labelencoder_X_1 = LabelEncoder()
y = labelencoder_X_1.fit_transform(y_categorical)
```

```python
# We can see that # 19, 20 and 21 are "0" within the array that goes from 0:29
# Therefore, the sanity check confirms that we have done the encoding correctly
display(y[:30])
print(y.shape)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
# Due to the nature of ANNs we must scale our features before fitting a Neural Network Model
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Import packages required for an ANN
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
```

```python
#Ensure recomputability
from numpy.random import seed
seed(4)
from tensorflow import set_random_seed
set_random_seed(9)

#Create the neural network
ann = Sequential()

#Add input and 1st hidden layer with a dropout of 20% prob
ann.add(Dense(16, activation='relu', kernel_initializer='uniform', input_dim=(30)))
ann.add(Dropout(0.2))
#2nd hidden layer with a dropout of 20% prob
ann.add(Dense(16, activation='relu', kernel_initializer='uniform'))
ann.add(Dropout(0.2))
#Output layer, sigmoid activation function for binary output
ann.add(Dense(1, activation='sigmoid', kernel_initializer='uniform'))

#Compile the model
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
ann.fit(X_train, y_train, epochs=150, batch_size=100)
```

```python

prediction=ann.predict(X_test)

y_pred=[1 if x == True else 0 for [x] in prediction >= 0.5]

print("Accuracy score: ",accuracy_score(y_test, y_pred))
```

```python
#Let's tune the drop out parameter
from keras.wrappers.scikit_learn import KerasClassifier

#Create the model for KerasClassifier
def create_model(dropout_rate=0.0):
            ann  = Sequential()
            ann.add(Dense(16, activation='relu', kernel_initializer='uniform', input_dim=(30)))
            ann.add(Dropout(dropout_rate))
            ann.add(Dense(16, activation='relu', kernel_initializer='uniform'))
            ann.add(Dropout(dropout_rate))
            ann.add(Dense(1, activation='sigmoid', kernel_initializer='uniform'))
            ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
            return ann

#Set up our pipeline for gridsearchCV
from sklearn.pipeline import Pipeline 
from sklearn.model_selection import GridSearchCV

#Set for Reproducibility 
from numpy.random import seed
seed(9)

#Setting up the estimator for gridsearchcv
model = KerasClassifier(build_fn=create_model, verbose=10)

#Defining paramater grid
#Since accuracy was extremely high for drop_out rate of 20%, we'll select our paramater grid around 0.2
dropout_rate = [0.19, 0.20, 0.21]
param_grid = dict(dropout_rate = dropout_rate)

#Using CV=5
grid_search = GridSearchCV(estimator = model, param_grid = param_grid, cv=5)
#Decrease number of epochs to 10 for computational feasibility
grid_search.fit(X_train, y_train, epochs=10, batch_size=100)
print("Best parameters: {}".format(grid_search.best_params_))
print("Best cross-validation score: {:.4f}".format(grid_search.best_score_))

#Compute mean accuracy scores for each gridsearched parameter
means = grid_search.cv_results_['mean_test_score']
stds = grid_search.cv_results_['std_test_score']
params = grid_search.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
```

```python
# Predicting the Test set results with the best parameters from tuning
prediction = grid_search.predict(np.array(X_test))
print("Test accuracy score: {:.2f}".format(grid_search.score(X_test,y_test)))
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

y_pred=[1 if x == True else 0 for [x] in prediction >= 0.5]

#Accuracy: as shown above!
print("Accuracy score: ",accuracy_score(y_test, y_pred))

#Computing Confusion Matrix
confusion = confusion_matrix(y_test, y_pred)
print("Confusion matrix:\n{}".format(confusion))

#Precision, Recall and F1 score 
print(classification_report(y_test, y_pred))
```
